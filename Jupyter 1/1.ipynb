```python
import pandas as pd
from datetime import datetime
from docxtpl import DocxTemplate
from gtts import gTTS
```

```python
students = pd.read_excel("Jupyter1 - Students.xlsx")
```

```python
data = []
for index,row in students.iterrows():
    average = (row["math"] + row["english"] + row["science"]) / 3
    average = round(average,2)
    if average >= 15:
        status = "Honor"
    elif average >= 10:
        status = "Aprobado"
    elif average <10:
        status = "Reprobado"
    result = {
        "name": row["name"],
        "math": row["math"],
        "english": row["english"],
        "science": row["science"],
        "average": average,
        "status": status
    }
    data.append(result)
    data_df = pd.DataFrame(data)
    data_df.to_excel('notes_resume.xlsx')
```

```python
report_card_info = pd.read_excel("notes_resume.xlsx")
report_card = DocxTemplate("Report_card.docx")
```

```python
for index,row in report_card_info.iterrows():
    student_data = {
        "name": row["name"],
        "math": row["math"],
        "english": row["english"],
        "science": row["science"],
        "average": row["average"],
        "status": row["status"]
    }
    report_card.render(student_data)
    report_card.save(f"Boletin - {row["name"]}.docx")

    text = f"Epale pue, su hijo, {row['name']} anda con un promedio de {row['average']}, y ahorita anda pero bueno pue, {row['status']}"
    tts = gTTS(text=text,lang="es")
    filename = f"Audio - {row['name']}.mp3"
    tts.save(filename)
```
