```python
from docxtpl import DocxTemplate, InlineImage, RichText
from docx.shared import Mm, Inches, Pt, RGBColor
import os
```

```python
doc = DocxTemplate("Template.docx")
imagen = InlineImage(doc, 'image.jpg', width=Mm(50))
```

```python
context = {"img": imagen}
doc.render(context)
doc.save("Template1.docx")
```

```python

table = doc.add_table(rows=2, cols=3)
hdr_cells = table.rows[0].cells
hdr_cells[0].text = 'Info1'
hdr_cells[1].text = 'Info2'
hdr_cells[2].text = 'Info3'

row_cells = table.rows[1].cells
row_cells[0].text = '{% for a in list %}{{ a.nombre }}'
row_cells[1].text = '{{ a.ranNum }}'
row_cells[2].text = '{{ a.spot }}{% endfor %}'

doc.save("Template_base.docx")
doc2 = DocxTemplate("Template_base.docx")

list = [
    {'name': 'Yomi', 'RanNum': 2, 'spot': 5},
    {'name': 'Tomo', 'RanNum': 5, 'spot': 1},
    {'name': 'Chiyo', 'RanNum': 6, 'spot': 6},
]
context2 = {'list': list}
doc2.render(context2)
doc2.save("Template2.docx")
```
