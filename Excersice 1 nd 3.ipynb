```python
from gtts import gTTS
import os
from IPython.display import Audio, display
```

```python
tts_text1 = "Este texto está en español de españa y es rapido"
tts1 = gTTS(text=tts_text1,lang="es",tld="es",slow=False)
fname1 = "Audio1.mp3"
tts1.save(fname1)

display(Audio(fname1))
```

```python
tts_text2 = "Este texto está en español de españa y es lento"
tts2 = gTTS(text=tts_text2,lang="es",tld="es",slow=True)
fname2 = "Audio1.mp3"
tts2.save(fname2)

display(Audio(fname2))
```
