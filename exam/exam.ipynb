```python
all_sheets_data = pd.read_excel("Libro1.xlsx", sheet_name=None)
doc = DocxTemplate("Gasto de Deprtamento.docx")

for sheet_name, df in all_sheets_data.items():
    print(df.head())
    total_spendings=0
    spendings = []
    first_row = df.iloc[0]
    department_name = first_row["department"]
    budget = first_row["budget"]
    for index,row in df.iterrows():
        total_spendings += row["spendings"]
        spendings.append(row["spendings"])
        doc.add_paragraph(department_name)
    
```

```python

table = doc.add_table(rows=len(spendings), cols=1)
hdr_cells = table.rows[0].cells
hdr_cells[0].text = 'Spendings'
row_cells = table.rows[1].cells
row_cells[0].text = '{%for a in spendings%}{{a}}'
doc.render
doc.save('template.docx')
```
