```python
import pandas as pd
from datetime import datetime
from docxtpl import DocxTemplate
```

```python
doc = DocxTemplate("Certificate.docx")
```

```python
course = "Python"
date = "10/11/2025"
signature = "Luis A."
```

```python
Data = {
    "course": course,
    "date": date,
    "signature": signature
}
```

```python
print(Data)
```

```python
%pip install openpyxl
ss = pd.read_excel("Students.xlsx")
```

```python
ss
```

```python
for index,row in ss.iterrows():
    Student_data = {
        "name": row["name"],
        "student_mail": row["student_mail"]
    }
    doc.render({**Data, **Student_data})
    doc.save(f"Certificado - {row['name']}.docx")
```
