```python
import pandas as pd
from datetime import datetime
from docxtpl import DocxTemplate
```

```python
doc = DocxTemplate("FACTURA.docx")
```

```python
%pip install openpyxl -q
ss = pd.read_excel("Clients.xlsx")
```

```python
vendor = "GipShop"
date = datetime.today().strftime("[%d/%m/%Y]")
```

```python
VendorData = {
    "vendor": vendor,
    "date": date
}
```

```python
print(VendorData)
```

```python
for index,row in ss.iterrows():
    total = row["product_price1"] + row["product_price2"]
    ClientData = {
        "name": row["name"],
        "product1": row["product1"],
        "product_price1": row["product_price1"],
        "product2": row["product2"],
        "product_price2": row["product_price2"],
        "total": total
    }
    doc.render({**VendorData, **ClientData})
    doc.save(f"FACTURA - {row['name']}.docx")
```
